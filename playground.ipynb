```python
class Queue():
    def __init__(self):
        self.queue = []
    
    # Adds new values into the queue
    def enqueue(self, val):
        self.queue.append(val)
      
    # Removes the first values inserted into the queue  
    def dequeue(self):
        self.queue.pop(0)
        
if __name__ == "__main__":
    q = Queue()
    q.enqueue(10)
    q.enqueue(20)
    q.enqueue(30)
    q.enqueue(40)
    q.enqueue(50)
    
    print("Values after insertion :", q.queue)
    q.dequeue()
    q.dequeue()
    print("Values after deletion :", q.queue)
    
```

```python
class MergeSortedArrays:
    def merge_sorted_arrays(self, arrA, arrB):
        return sorted(set(arrA + arrB))
    
    def main(self):
        a = [1, 5, 9, 15, 9, 4, 6, 10, 5, 11]
        b = [10, 12, 1, 3, 2, 7, 9, 14, 8, 13]
        c = self.merge_sorted_arrays(a, b)
        
        print(c)
        
if __name__ == "__main__":
    MergeSortedArrays().main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
        
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        self.size += 1
        
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1
        
    def insert(self, index, data):
        if not 0 <= index <= self.size:
            raise IndexError("Index out of bounds")
        
        if index == 0:
            self.prepend(data)
        else:
            new_node = Node(data)
            current = self.head
            for _ in range(index - 1):
                current = current.next
            new_node = current.next
            current.next = new_node
    
    def delete(self, data):
        if self.head is None:
            raise ValueError("List is empty")  
        
        if self.head.data == data:
            self.head = self.head.next
            self.size -= 1
            return
        
        current = self.head
        while current.next and current.next.data != data:
            current = current.next
        
        if current.next is None:
            raise ValueError("Data not found")
        
        current.next = current.next.next
        self.size -= 1
    
    def search(self, data):
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
            
    def get_size(self):
        return self.size
        
    def traverse_linked_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print(None)
        
def main():
    llist = LinkedList()
        
    print("===== Append Operation =====")
    llist.append(1)
    llist.append(2)
    llist.append(3)
    llist.traverse_linked_list()
    
    print("\n=== Prepend Operation ===")
    llist.prepend(0)
    llist.traverse_linked_list()
    
    print("\n=== Insert Operation ===")
    llist.insert(2, 5) # Insert 5 at index 2
    llist.traverse_linked_list()
    
    print("\n=== Delete Operation ===")
    llist.delete(2)  # Delete node with data 2
    llist.traverse_linked_list()
    
    print("\n=== Search Operation ===")
    print("Is 3 in the list?", llist.search(3))
    print("Is 10 in the list?", llist.search(10))
    
    
    

if __name__ == "__main__":
    main()
```
